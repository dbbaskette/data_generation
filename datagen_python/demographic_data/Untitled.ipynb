```python
import pandas
from pandas import *
```

```python
df = read_csv('locations_partitions.csv')
```

```python
df.loc[:, 'diff'] = 0

for i in df.index:
    if i == 0:
        df.loc[i, 'diff'] = df.loc[i, 'partition']
    else:
        df.loc[i, 'diff'] = df.loc[i, 'partition'] - df.loc[i-1, 'partition']
```

```python
df.sort('diff', ascending = False)
```
